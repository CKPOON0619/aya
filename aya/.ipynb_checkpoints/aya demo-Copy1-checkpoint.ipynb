```python
import tensorflow as tf
import numpy as np
import itertools
import os
```

```python
#Clearing all
tf.reset_default_graph()
#Opening new directory
from datetime import datetime
now = datetime.now()
logdir = 'C:/Users/kin.poon/Desktop/tf_logs' + now.strftime("%Y%m%d-%H%M%S") + "/"
print('TensorBoard CMD:tensorboard --logdir='+logdir)
```

```python
learning_rate = 0.001
epochs = 100
batch_size = 50
```

```python
import csv

def generator():
    Reader = csv.reader(open('./test inputs/attritionSMALL.txt', newline='\r\n'), delimiter=',', quotechar='|',)
    header=next(Reader)
    data_batch=list(itertools.islice(Reader, 0, batch_size))
    while len(data_batch)>0:
      try:
        yield data_batch
      except tf.errors.OutOfRangeError:
        break
        
    
```

# Training

```python
dataReader = csv.reader(open('./test inputs/attritionSMALL.txt', newline='\r\n'), delimiter=',', quotechar='|',)
header=next(dataReader)
training_data=list(dataReader)
training_inputs=np.array([r[:51] for r in training_data],dtype=np.float32)
training_labels=np.array([r[51:52] for r in training_data],dtype=np.float32)
```

```python
#inputs_=batch[:,:51]
#labels_=batch[:,51:52]

inputs_ = tf.placeholder(tf.float32, (None, 51), name='inputs')
labels_ = tf.placeholder(tf.float32, (None, 1), name='labels')

inputs = tf.data.Dataset().from_tensor_slices(inputs_).batch(batch_size)
labels = tf.data.Dataset().from_tensor_slices(labels_).batch(batch_size)

inputs_iter = inputs.make_initializable_iterator()
labels_iter = labels.make_initializable_iterator()

inp=inputs_iter.get_next()
lbl=labels_iter.get_next()

input_norm=tf.layers.batch_normalization(inp,axis=1,trainable=True, training=True)
layer1=tf.layers.dense(input_norm,units=500,activation=tf.nn.sigmoid)
layer1_norm=tf.layers.batch_normalization(layer1,axis=1,trainable=True, training=True)
layer2=tf.layers.dense(layer1_norm,units=200,activation=tf.nn.sigmoid)
layer2_norm=tf.layers.batch_normalization(layer2,axis=1,trainable=True, training=True)
layer3=tf.layers.dense(layer2_norm,units=50,activation=tf.nn.sigmoid)
layer4=tf.layers.dense(layer3,units=1,activation=None)
output=tf.nn.sigmoid(layer4,name='output')
loss=tf.nn.sigmoid_cross_entropy_with_logits(labels=lbl, logits=layer4)
cost = tf.reduce_mean(loss)
update_ops = tf.get_collection(tf.GraphKeys.UPDATE_OPS)
with tf.control_dependencies(update_ops):
    optimiser=tf.train.AdamOptimizer(learning_rate)
    opt=optimiser.minimize(loss)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
OutOfRangeError=False
for e in range(epochs):
    sess.run(inputs_iter.initializer, feed_dict={ inputs_: training_inputs})
    sess.run(labels_iter.initializer, feed_dict={ labels_: training_labels})
    try:
        #_,batch_cost=sess.run([opt,cost],feed_dict={inputs_:training_inputs,labels_:training_labels})
        _,batch_cost=sess.run([opt,cost])
        #print("Running epochs {}, batch {}".format(e+1,ii+1))
    except tf.errors.OutOfRangeError:
        OutOfRangeError=True
        break
    if(OutOfRangeError):
        break
    print("Epoch: {}/{}...".format(e+1, epochs),
          "Training loss: {}".format(batch_cost))

builder = tf.saved_model.builder.SavedModelBuilder(logdir)
builder.add_meta_graph_and_variables(sess,[tf.saved_model.tag_constants.SERVING],strip_default_attrs=True)
builder.save()
```

```python
#To show the tensorBoard
from IPython.display import clear_output, Image, display, HTML

def strip_consts(graph_def, max_const_size=32):
    """Strip large constant values from graph_def."""
    strip_def = tf.GraphDef()
    for n0 in graph_def.node:
        n = strip_def.node.add() 
        n.MergeFrom(n0)
        if n.op == 'Const':
            tensor = n.attr['value'].tensor
            size = len(tensor.tensor_content)
            if size > max_const_size:
                tensor.tensor_content = "<stripped %d bytes>"%size
    return strip_def

def show_graph(graph_def, max_const_size=32):
    """Visualize TensorFlow graph."""
    if hasattr(graph_def, 'as_graph_def'):
        graph_def = graph_def.as_graph_def()
    strip_def = strip_consts(graph_def, max_const_size=max_const_size)
    code = """
        <script>
          function load() {{
            document.getElementById("{id}").pbtxt = {data};
          }}
        </script>
        <link rel="import" href="https://tensorboard.appspot.com/tf-graph-basic.build.html" onload=load()>
        <div style="height:600px">
          <tf-graph-basic id="{id}"></tf-graph-basic>
        </div>
    """.format(data=repr(str(strip_def)), id='graph'+str(np.random.rand()))

    iframe = """
        <iframe seamless style="width:1200px;height:620px;border:0" srcdoc="{}"></iframe>
    """.format(code.replace('"', '&quot;'))
    display(HTML(iframe))
show_graph(tf.get_default_graph().as_graph_def())
```
